````python
%time x = np.linspace(0, 2*np.pi, 1000000)
````

````python
%%timeit

for i in range(10000):
    x[i] = np.sin(x[i])
````

````python
%%timeit

import concurrent.futures

def f(x):
    return np.sin(x)

with concurrent.futures.ProcessPoolExecutor() as executor:
    x = np.linspace(0, 2*np.pi, 10000)
    executor.map(f, x)
````

````python
def cubic_solver(a, b, c, d):
    """Solve a cubic equation.

    Parameters
    ----------
    a : float
        Coefficient of x^3 term
    b : float
        Coefficient of x^2 term
    c : float
        Coefficient of x term
    d : float
        Coefficient of constant term

    Returns
    -------
    x : array_like
        Real roots of cubic equation
    """
    # Make a function that returns the fibonacci sequence
    x = np.roots([a, b, c, d])
    return x
````

````python
# Make a function that returns the factorial of a number

def factorial(n):
    """
    Make a function that returns the factorial of a number
    :param n: The number
    :return: The factorial of n
    """
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
````

## Upload your code to GitHub

You can upload your code to GitHub directly from PyCharm, without having to use the command line. You can find instructions on how to do that at [https://www.jetbrains.com/help/pycharm/set-up-a-git-repository.html](https://www.jetbrains.com/help/pycharm/set-up-a-git-repository.html). But we will introduce how to upload your code from PyCharm to GitHub in the next steps:

1. Create a new project in PyCharm
2. Make some work in your project (create code files, directories, notebooks, etc)
3. Go to VCS menu and select Create a new Git repository
4. Go to the VCS menu and select Import into Version Control -> Share Project on GitHub
5. Enter your GitHub credentials
6. Enter a name for your repository
7. Click on Share
8. Go to your GitHub account and check that your repository has been created
